#Decisão 1:
Para cada Atributo decidir entre: 
*   e: excluir
*   n: atributo do tipo numérico (não será excluído)
*   c: atributo do tipo categórico (não será excluído)  
*   t: target, alvo da investigação  
Formato da informação:
lista de tuplas com: (nome atributo, decisão)

```python
arquivo = "decisao_base.xlsx"
caminho = pasta + arquivo
decisoes = pd.read_excel(caminho)
decisoes.head(2)
```

Determinando tipo de cada atributo  segundo critério da tabela de decisão

```python
print("antes: ", df.shape)
excluir = []
for index in range(0, len(decisoes)):
    decisao_atributo = decisoes["decisao"].iloc[index] #decisão tomada
    nomeAtributo = df.columns[index] #nome do atributo relacionado a decisão
    if decisao_atributo == 'c':
        df[nomeAtributo] = df[nomeAtributo].astype('category') #mudando tipo do dado
    elif decisao_atributo == 'n':
        df[nomeAtributo] = df[nomeAtributo].astype('float64') #mudando tipo do dado
    elif decisao_atributo == 'e':
        excluir.append(nomeAtributo) #adicionando atributos que serão exluídos.
    elif decisao_atributo == 't':
        excluir.append(nomeAtributo)
        df["target"] = df[nomeAtributo].astype('category')
    else:  
        print("Atributo não previsto: ", nomeAtributo)
df = df.drop(columns = excluir) #excluir colunas
print("depois: ", df.shape)
```

# Conhecendo as variáveis

```python
tipos = df.dtypes
atributos = df.columns

for i in range(len(atributos)):
    atributo = atributos[i]
    tipo = tipos[i]

    print('######################################################')
    print('### ATRIBUTO: %s (%s)' % (atributo, tipo) )
    print('######################################################')
    print("NaN: ", df[atributo].isnull().sum()/df.shape[0])
    if(tipo == 'int64' or tipo == 'float64'):        
        print(df[atributo].describe())
        #Histograma
        df[atributo].hist(bins=10)
        plt.show()
        #Boxplot
        df.boxplot(column=[atributo])
        plt.show()
        
    elif(tipo == 'category'):
        #Descrição em números
        print('\nDistribuição por frequência')
        print(df[atributo].value_counts())
        #s.value_counts(normalize=True)
        #s.value_counts(dropna=False) see Na values
        #self, normalize=False, sort=True, ascending=False, bins=None, dropna=True
        print('\nDistribuição por frequência %')
        print(df[atributo].value_counts(normalize=True))
        #Criando gráfico de barras
        print('\nGráfico de barras')
        df[atributo].value_counts().plot(kind='bar')
        plt.show()
    else:
        print('!!!!!!!!!!!!!!!!!! tipo não previsto')
```

# Conhecendo as variáveis em relação ao rótulo

```python
tipos = df.dtypes
atributos = df.columns
group0 = df[df['target'] == "Sim"]
group1 = df[df['target'] == "Não"]


for i in range(len(atributos)):
    atributo = atributos[i]
    tipo = tipos[i]

    print('######################################################')
    print('### ATRIBUTO: %s (%s)' % (atributo, tipo) )
    print('######################################################')
    if(tipo == 'int64' or tipo == 'float64'):        
        grouped = df.groupby('target')
        # create a box plot of the attribute in relation to the target variable
        grouped.boxplot(column=atributo)
        # set plot labels
        plt.show()
        
        #df.boxplot(x = column=['target', atributo])
        #plt.show()
        
    elif(tipo == 'category'):
        #Descrição em números
        # create histograms for each group
        plt.hist(group0[atributo], alpha=0.5, label='Não')
        plt.hist(group1[atributo], alpha=0.5, label='Sim')
        plt.xlabel(atributo)
        plt.ylabel('Frequency')
        plt.legend(loc='upper right')

        # show the plot
        plt.show()
    else:
        print('!!!!!!!!!!!!!!!!!! tipo não previsto')
```
